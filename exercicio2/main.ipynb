```python
import pandas as pd
```

```python
df_funcionarios = pd.read_excel("funcionarios.xlsx")
df_departamentos = pd.read_excel("departamentos.xlsx")


display(df_funcionarios, "\n\n\n", df_departamentos)
```

```python
df = pd.merge(df_funcionarios, df_departamentos, left_on="Worker ID", right_on="Worker Ref ID")

display(df)
```

```python
df_final = df.groupby("Worker Title")["Salary"].sum()

display(df_final)
```

```python
df_final = df_final.sort_values(ascending=False)

display(df_final.iloc[:2])
```
