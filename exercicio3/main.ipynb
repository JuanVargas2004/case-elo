```python
import pandas as pd
```

```python
df = pd.read_excel("retencao.xlsx")

display(df)
```

```python
df_dezembro = df[   (df["date"] >= "2022-12-01")    &   (df["date"] < "2023-01-01")  ]
df_janeiro = df[   (df["date"] >= "2023-01-01")    &   (df["date"] < "2023-02-01")  ]

display(df_dezembro, "\n\n\n\n", df_janeiro)
```

```python
qntd_dezembro = df_dezembro.groupby(by="account_id")["user_id"].nunique()
qntd_janeiro = df_janeiro.groupby(by="account_id")["user_id"].nunique()


retencao = (qntd_dezembro/qntd_janeiro)*100

print(retencao.dropna())
```
